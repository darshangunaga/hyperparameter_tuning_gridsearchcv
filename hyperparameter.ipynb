```python
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris = load_iris()
```

```python
iris.keys()
```

```python
df = pd.DataFrame(iris.data,columns = iris.feature_names)
```

```python
df.head()
```

```python
df['target']= iris.target
```

```python
df.head()
```

```python
inputs = df.drop(df.target)
target = df['target']
```

```python
from sklearn import svm
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(iris.data,iris.target, test_size=0.3 )
```

```python
model = svm.SVC(kernel="rbf",C= 30,gamma='auto')
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
from sklearn.model_selection import KFold,cross_val_score
```

```python
cross_val_score(svm.SVC(kernel="rbf",C=30,gamma="auto"),iris.data,iris.target)
```

```python
kernels = ['rbf','linear']
cs = [10,20,30,40]
length = len(kernels)
length2 = len(cs)
for i in range (0,length):
    for j in range(0,length2):
        score =cross_val_score(svm.SVC(kernel=kernels[i],C=cs[j],gamma = "auto"), iris.data, iris.target)
        print(kernels[i],cs[j],score.mean())
        
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
gs = GridSearchCV(svm.SVC(gamma="auto"), {
                'C' : [1,10,20],
                'kernel' : ['linear','rbf']
},return_train_score = False)
gs.fit(iris.data,iris.target)
gs.cv_results_
df1 =  pd.DataFrame(gs.cv_results_) 
df1
```
